```python
main_data['Item_Fat_Content'] = main_data['Item_Fat_Content'].replace({'LF' : 'Low Fat', 'reg' : 'Regular', 'low fat': 'Low Fat'})
main_data['Item_Fat_Content'].value_counts()
```

```python
main_data.loc[main_data['combined_item_type']=="Non-Consumable",'Item_Fat_Content'] = "Non-Edible"
main_data['Item_Fat_Content'].value_counts()
```

```python
#list of columns for the model
lst_model = ['Item_Fat_Content', 'Item_MRP',
       'Item_Outlet_Sales', 'Item_Type', 'Item_Visibility', 'Item_Weight',
        'Outlet_Location_Type', 'Outlet_Size','Outlet_Identifier',
       'Outlet_Type', 'source', 'Oulet_age', 'combined_item_type']
model_data = main_data[lst_model]
```

```python
from sklearn.preprocessing import LabelEncoder
cat_columns = ("Item_Fat_Content", "combined_item_type","Outlet_Identifier","Outlet_Location_Type","Outlet_Size", "Outlet_Type")
for col in cat_columns:
    label = LabelEncoder()
    label.fit(list(main_data[col].values))
    main_data[col] = label.transform(list(main_data[col].values))
```

```python
main_data.head()
```

```python
train_d = main_data[main_data['source'] == 'train']
train_d = train_d.drop(columns = 'source')
test_d = main_data[main_data['source'] == 'test']
test_d = test_d.drop(columns = 'source')
```

```python
#list of columns for the model
lst_model = ['Item_Fat_Content', 'Item_MRP', 'Item_Visibility', 'Item_Weight',
        'Outlet_Location_Type', 'Outlet_Size','Outlet_Identifier',
       'Outlet_Type', 'Oulet_age', 'combined_item_type']
train_dx =  train_d[lst_model]
train_dy = train_d['Item_Outlet_Sales']
test_dx =  test_d[lst_model]
```

```python
from sklearn.ensemble import RandomForestRegressor
random_regressor = RandomForestRegressor(n_estimators=200,max_depth=5, min_samples_leaf=100,n_jobs=4)
random_regressor.fit(train_dx, train_dy)
forest_predictions = random_regressor.predict(train_dx)
print("Mean Absolute Error : " + str(mean_absolute_error(train_dy,forest_predictions)))
```

```python
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor

my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
my_model.fit(train_dx,train_dy)
predictions = my_model.predict(train_dx)

from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error : " + str(mean_absolute_error(train_dy,predictions)))
```

```python
#checking the model performance using k-fold cross validation
accuracies = cross_val_score(estimator= my_model, X= train_dx, y = train_dy, cv =10)
print(accuracies.mean())
print(accuracies.std())
```
